```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset_path = '/content/drive/MyDrive/archive/At_India_2011.csv'
At_India_2011 = pd.read_csv(dataset_path)
dataset_path = '/content/drive/MyDrive/archive/In_India.csv'
In_India = pd.read_csv(dataset_path)
```

```python
# Calculate unemployment rate for 2011
At_India_2011['UnemploymentRate'] = ((At_India_2011['Estimated_Unemployment_Rate'] - At_India_2011['Estimated_Employment']) / At_India_2011['Estimated_Rate(%)']) * 100
```

```python
# Calculate unemployment rate for 2019
In_India['UnemploymentRate'] = (In_India['Unemployment']/ In_India['Estimated_Rate(%)']) * 100
```

```python
plt.figure(figsize=(14, 8))
```

```python
# Plotting graphs for 2011
plt.subplot(1, 2, 1)
for region in At_India_2011['Region'].unique():
    region_data = At_India_2011[At_India_2011['Region'] == region]
    plt.plot(region_data['Date_Column'], region_data['UnemploymentRate'], marker='o', label=region)
plt.title('Unemployment Rate up to 2011 -2020')
plt.xlabel('Date Column')
plt.ylabel('Unemployment Rate (%)')
plt.legend()
plt.xticks(rotation=45)

# Plotting graphs for 2019
plt.subplot(1, 2, 2)
for region in In_India['Region'].unique():
    region_data = In_India[In_India['Region'] == region]
    plt.plot(region_data['Date_Column'], region_data['UnemploymentRate'], marker='o', label=region)
plt.title('Unemployment  in india')
plt.xlabel('Date Column')
plt.ylabel('Unemployment Rate (%)')
plt.legend(loc='upper left')  # Adjust legend placement
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
